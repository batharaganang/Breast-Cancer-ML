```python
X_train, X_test, y_train, y_test = train_df.drop(['sampleID','target'], axis=1),test_df.drop(['sampleID','target'], axis=1),train_df['target'],test_df['target']
```

```python
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

# Modelling

    When average=None, precision and recall scores are calculated for each class separately, and an array of scores is returned.
    When average='macro', the precision and recall scores are calculated by averaging the scores for each class.
    When average='micro', the precision and recall scores are calculated globally by counting the total true positives, false positives, and false negatives across all classes.
    When average='weighted', the precision and recall scores are calculated as the weighted average, weighted by the number of samples in each class.

```python
# Create an empty DataFrame with column names
columns = ['model', 'scaler', 'mean_train_accuracy', 'test_accuracy', 'precision', 'recall', 'f1_score', 'cm']
base_model_df = pd.DataFrame(columns=columns)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import StratifiedKFold

# 5 fold stratified cross validation
fold = 5
stratified_kfold = StratifiedKFold(n_splits=fold, shuffle=True, random_state=42)
```

# Base Modelling

```python
# Create an Logistic Regression classifier
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# Perform cross-validation
scores = cross_val_score(model, X_train, y_train, cv=stratified_kfold, scoring='accuracy')

# Print the cross-validation scores
print("Cross-Validation Scores:", scores)
print("Mean Accuracy:", scores.mean())

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate the multilabel confusion matrix
cm = multilabel_confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Calculate weighted-averaged precision and recall scores
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
print("Weighted-averaged Precision:", precision)
print("Weighted-averaged Recall:", recall)

# Calculate F1 score
f1 = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1)

# Create a new row as a dictionary
new_row = {'model': 'LogisticRegression', 'scaler': 'No Scaler', 'mean_train_accuracy': scores.mean(), 'test_accuracy': accuracy, 'precision': precision, 'recall': recall, 'f1_score': f1, 'cm': cm}
# Append the new row to the DataFrame
base_model_df = base_model_df.append(new_row, ignore_index=True)
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [4, 5, 6],
    'learning_rate': [0.3, 0.1],
    'min_child_weight': [1, 5, 10]
}

# Create an XGBoost classifier
import xgboost as xgb
model = xgb.XGBClassifier()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=model, param_grid=param_grid,
                           cv=stratified_kfold,
                           scoring='accuracy',
                           verbose=2,
                           n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train_fs, y_train)

# Get the grid search CV results
cv_results = grid_search.cv_results_

# Create an empty DataFrame with column names
columns = ['cv_accuracy','learning_rate','max_depth','min_child_weight','n_estimators']
base_model_df = pd.DataFrame(columns=columns)

# Print the CV results
for mean_score, params in zip(cv_results["mean_test_score"], cv_results["params"]):
    # Create a new row as a dictionary
    new_row = {'cv_accuracy': mean_score, 'learning_rate': params['learning_rate'], 'max_depth': params['max_depth'], 'min_child_weight': params['min_child_weight'], 'n_estimators': params['n_estimators']}
    # Append the new row to the DataFrame
    base_model_df = base_model_df.append(new_row, ignore_index=True)
    print(mean_score, params)

# Get the best parameters
best_params = grid_search.best_params_
print('Best params = ', best_params)
```
